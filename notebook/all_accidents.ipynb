THE PROCESS OF CONVERTING OBJECT COLUMNS FROM AND OBJ TO AN INT & REMOVING REPLACING THE ',' IN THE COLUMNS

```python
total_acc_2022 = total_acc_2022.replace(',','', regex=True)
```

```python
total_acc_2022['Bus (Seats for 9-15 People, including driver)'] = total_acc_2022['Bus (Seats for 9-15 People, including driver)'].astype(str).astype(int)
```

```python
total_acc_2022.dtypes
```

```python
top_10_in_2022 = total_acc_2022.nlargest(10, "Total")
```
