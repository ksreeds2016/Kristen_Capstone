```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
total_acc_2022_by_vehicle = pd.read_csv('C:\\Users\\ksree\\Documents\\dda9\\PYTHON\\PROJECTS\\Kristen_Capstone\\data\\2022_all_accidents_by_mc.csv')
```

```python
total_acc_2022_by_vehicle.head()
```

```python
total_acc_2022_by_vehicle[total_acc_2022_by_vehicle.columns[2:]]=total_acc_2022_by_vehicle[total_acc_2022_by_vehicle.columns[2:]].replace(',','',regex=True).astype('int64')
```

```python
vehicle_highest_2022 = total_acc_2022_by_vehicle.nlargest(3, "Sub Total")
```

```python
vehicle_highest_2022
```

Tractor/Semi-trailer has the highest accidents in 2022 at 83k

```python
vehicle_lowest_2022 = total_acc_2022_by_vehicle.nsmallest(3, "Sub Total")
```

```python
vehicle_lowest_2022
```

Passenger Car has the lowest at 1, followed by light truck at 19, and tractor/triple at 89 for 2022
